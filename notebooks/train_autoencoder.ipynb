```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

columns = df.columns
df[columns] = scaler.fit_transform(df[columns])
```

```python
df.describe()
```

```python
split_date = "2008-11-10 23:00:00"

df_train = df.loc[:split_date]
df_test =  df.loc[split_date:]
```

```python
df_train.describe()
```

```python
import pickle

with open("scaler.pkl","wb+") as fout:
    fout.write(pickle.dumps(scaler))
```

```python
df_train.to_hdf("../input/hdfs-dataset/train.hd5","train")
df_test.to_hdf("../input/hdfs-dataset/test.hd5","test")
```

```python
from keras.layers import Input, Dense, Dropout
from keras.models import Model
from keras.callbacks import ReduceLROnPlateau


from keras import regularizers

# this is the size of our encoded representations
encoding_layers = [32,16,8]

decoding_layers = reversed(encoding_layers[:-1])

# this is our input placeholder
L = Input(shape=df.columns.shape,name="encode_input")
inputs = L

kwargs = dict(use_bias=True)#, kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))
for ilayer,dim in enumerate(encoding_layers):
    L = Dense(dim,activation="relu",name="encode_%d" % ilayer, **kwargs)(L)
    L = Dropout(0.4,name="do_%d" % ilayer)(L)
    
outputs = L   
encoder = Model(inputs=inputs,outputs=outputs,name="encoder")

# this is our input placeholder
L = Input((encoding_layers[-1],),name="decode_input")
inputs = L

for ilayer,dim in enumerate(decoding_layers):
    L = Dense(dim,activation="relu",name="decode_%d" % ilayer, **kwargs)(L)
    L = Dropout(0.4,name="do_%d" % ilayer)(L)

L = Dense(df.columns.shape[0],activation=None,name="decoded")(L)


outputs = L   
decoder = Model(inputs=inputs,outputs=outputs,name="decoder")



```

```python
encoder.summary()
```

```python
decoder.summary()
```

```python
autoencoder = Model(inputs=encoder.inputs, outputs=decoder(encoder.outputs),name="autoencoder")
```

```python
autoencoder.summary()
```

```python
from keras.optimizers  import Adam
```

```python
optimizer = Adam(lr=1e-4)
```

```python
autoencoder.compile(optimizer=optimizer,loss="mse")
```

```python
history = autoencoder.fit(df_train,df_train,epochs=200,validation_split=0.2,callbacks=[ReduceLROnPlateau(patience=5)])
```

```python
train_true = df_train.values
test_true = df_test.values

train_pred = autoencoder.predict(train_true)
test_pred = autoencoder.predict(test_true)
```

```python
for col in range(55):
    plt.scatter(train_true[:,col],train_pred[:,col])
    plt.scatter(test_true[:,col],test_pred[:,col],alpha=0.5)
    plt.title(col)
    plt.show()
```
