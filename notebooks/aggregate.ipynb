```python
columns = ["TemplateId","Component","Level"]

df = pd.read_hdf("../input/hdfs-dataset/hdfs_structured_datetime.hd5",columns=columns)
```

```python
aggregated = None
dummy_cols = {}

for col in columns:
    print("Encoding " + col)
    encoded_col = pd.get_dummies(df[col])
    encoded_col.columns = [ col+"_"+str(x) for x in encoded_col.columns ]
    dummy_cols[col] = encoded_col.columns
    
    print("Aggregating " + col)
    aggregated_col = encoded_col.resample('T').sum()
    if aggregated is None:
        aggregated = aggregated_col
    else:
        print("Joining "+col)
        aggregated = aggregated.join(aggregated_col)

print("Done")
```
